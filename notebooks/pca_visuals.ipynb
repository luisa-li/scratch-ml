```python
from sklearn.datasets import load_iris
from pca import pca
import numpy as np
import matplotlib.pyplot as plt
```

```python
# get the data from sklearn dataasets, this part is referenced from Data Camp's tutorial for PCA
iris = load_iris()
data = iris.data
raw_labels = iris.target 
label = np.reshape(raw_labels,(-1,1))
concatenated_data = np.concatenate([data, label], axis=1)
transformed = pca(concatenated_data, 2) # reducing the dimensionality from 30 to 2 for visualization
```

```python
plt.scatter(transformed[:, 0], transformed[:, 1], c = label)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title("Principal component analysis of the Iris dataset")
plt.show()
```
